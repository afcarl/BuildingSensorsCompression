```python
%matplotlib inline
import pickle
import pandas as pd 
import sys
import numpy as np
from numpy import *
import matplotlib.pyplot as plt

A = pd.read_pickle('New150Days.pkl')
print 'input dataset:',shape(A)
```

```python
def epsilonRatio(data, epsilon = 0.5, iteration = 50):
    def getSample(threshhold):
        # result=[data[0]]
        cumulativeError = 0.0
        result = 1
        currentValue= data[0]
        for t in data:
            cumulativeError = cumulativeError + abs(t - currentValue)
            if cumulativeError>= threshhold:
                result = result + 1 #result.append(t)
                currentValue=t
                cumulativeError = 0.0
        return result

    total = data.size
    X = [epsilon * t for t in range(1, iteration+1)]
    Y = [1.0 * getSample(epsilon * t) / total for t in range(1, iteration+1)]
    plt.plot(X,Y)
    plt.xlabel('epsilon')
    plt.ylabel('ratio of samples')
    plt.show()
```
