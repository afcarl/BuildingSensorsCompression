```python
%matplotlib inline
import pickle
import pandas as pd 
import sys
import numpy as np
from numpy import *
from pprint import pprint
import matplotlib.pyplot as plt

A = pd.read_pickle('Compressed.pkl')
pprint(A)
```

```python
B = A['code']
B.sort(key=lambda tup: tup['time'])
Temperature_for_room=[]
for t in B:
   Temperature_for_room.append(t['value'])
pprint(Temperature_for_room[:20])
```

```python
def epsilonRatio(data, epsilon = 0.5, iteration = 50):
    def getSample(threshhold):
        # result=[data[0]]
        result = 1
        currentValue= data[0]
        for t in data:
            if abs( t-currentValue )>= threshhold:
                result = result + 1 #result.append(t)
                currentValue=t
        return result

    total = data.size
    X = [epsilon * t for t in range(1, iteration+1)]
    Y = [1.0 * getSample(epsilon * t) / total for t in range(1, iteration+1)]
    plt.plot(X,Y)
    plt.xlabel('epsilon')
    plt.ylabel('ratio of samples')
    plt.show()
```

```python
plt.plot(Temperature_for_room)
plt.xlabel('Time')
plt.ylabel('Temperature')
```

```python
epsilonRatio(np.array(Temperature_for_room), 0.1, 100)
```
